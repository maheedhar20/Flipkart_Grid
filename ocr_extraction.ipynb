```python
import pytesseract
from PIL import Image

# Load the image
image = Image.open('/content/test5.jpg')

# Extract text from the image using pytesseract
custom_config = r'--psm 11 --oem 3 '
text = pytesseract.image_to_string(image, config=custom_config )

# Print the extracted text
print(text)
```

```python
product_msg = HumanMessage(
    content=f"""
    Extract the following details from the text:
    'product_name','price','manufactured_date','expiry_date','quantity','manufactured_by','marketed_by','brand_name''ingredients'.
    Text from image:
    {text}
    """
)

prod_output = llm.invoke([product_msg])
print(prod_output.content)
```
